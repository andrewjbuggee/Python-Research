```python
xml_fname = 'LASICS-SPS_20230223T013121_NOAA20-Aug22-Sept22.xml'
df = pd.read_xml(xml_fname, xpath='/SPS_Plan/ScienceOpportunities/ScienceOpportunity')
exclude_columns = ['TargetName', 'TargetStartTime', 'TargetEndTime', 'ReferenceName']
df2 = (df.loc[:, ~df.columns.isin(exclude_columns)]).copy()
df2.columns = [df.ReferenceName[0]+'_StartTime', df.ReferenceName[0]+'_EndTime']
df2.ISS_StartTime = [x[0:-2]+'Z' for x in df2.ISS_StartTime]
df2.ISS_EndTime = [x[0:-2]+'Z' for x in df2.ISS_EndTime]
df2
```
